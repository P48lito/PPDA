** Set a variable X equal to the numerical features of the customers and a variable y equal to the "Yearly Amount Spent" column. **

```python
df.info()
```

```python
X = df[ ["Avg. Session Length", "Time on App", "Time on Website", "Length of Membership"] ]
X.info()
```

```python
y = df["Yearly Amount Spent"]
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# help(train_test_split)
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=101)
X_train.info()
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# y_p = a_0 + a_1 x_1 + a_2 * x_2 + a_3 * x_3 + a_4 * x_4
```

```python
# dir(model) # display attributes
```

```python
len(X_train)
```

```python
y_pred = model.predict(X_test)
# y_pred[:5]
```

```python
y_test.values[:5]
```

```python
f"{np.mean(np.abs(y_test - y_pred) / y_test * 100):.2f}%"
```

```python
plt.scatter(y_test, y_pred)
```

```python
print(f"MAE: {np.mean(np.abs(y_test-y_pred))}")
print(f"MSE: {np.mean(np.abs(y_test-y_pred)**2)}")
print(f"MAE: {np.sqrt(np.mean(np.abs(y_test-y_pred)**2))}")
```

```python
residuals = (y_pred-y_test)
sns.distplot(residuals)
```
