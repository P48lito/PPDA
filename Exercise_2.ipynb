Problem 1

Construct a generator which returns a sequence of prime numbers as a function, class and expression ().

a) as a function

```python
import math

def isprime(n):
  if n == 0:
    return False
  if n == 1:
    return False
  if n < 3:
    return True
  i = 2;
  while i <= math.sqrt(n):
    if n % i == 0:
      return False
    i += 1

  return True

def myprimef():
  print("Hello!")
  i = 1
  while True:
    if isprime(i):
      yield i
    i += 1

pg = myprimef()

first_10_prime = [ next(pg) for i in range(10)]
first_10_prime
```

b) as a class

```python
class MyPrimeC(object):
  def __init__(self):
    self._i = 0

  def __iter__(self):
    return self

  def __next__(self):
    self._i += 1
    while not isprime(self._i):
        self._i += 1
    return self._i

cg = MyPrimeC()


first_10_prime = [ next(cg) for i in range(10)]
first_10_prime
```

c) as an expression

```python
import itertools

# The second argument of an "islice" is about how many first prime numbers we want to display
tuple(itertools.islice((i for i in range(10000) if isprime(i) and i > 1), 11))
```

Problem 2

Fill in the Line class methods to accept coordinate as a pair of tuples and return the slope and distance of the line.

```python
import math

class Line(object):
  def __init__(self,point1,point2):
      """Initialize instance attributes with tuples (x1,y1)  and (x2,y2)
      """
      self._p1 = point1
      self._p2 = point2


  def distance(self):
      """Calculate the length of the segment (line)
      """

      p1 = self._p1
      p2 = self._p2
      return math.sqrt((p2[0] - p1[0])**2+(p2[1] - p1[1])**2)

  def slope(self):
      """ Return the slope of a line going through the ends ( the 'a' in y=ax+b)
      """
      p1 = self._p1
      p2 = self._p2
      return (p2[1] - p1[1]) / (p2[0]-p1[0])

p1 = (3,2)
p2 = (8,10)

li = Line(p1,p2)

print(li.distance())

print(li.slope())
```

Problem 3

Fill in the class

```python
class Cylinder(object):

  def __init__(self,height=1,radius=1):
      self._h = height
      self._r = radius

  def volume(self):
      return round(math.pi * self._r**2 * self._h, 2)

  def surface_area(self):
      return round(2 * math.pi * self._r**2 + 2 * math.pi * self._r * self._h, 1)

c = Cylinder(2,3)

print(c.volume())

print(c.surface_area())
```

Problem 4

Write a class that takes a filename as an argument, opens it and reads its content into a 2D matrix of values (an array of arrays: [[1,2],[3,4]]). The class should define a function info() which prints out statistical information about values in columns. Assume that the first row in the file specifies the column names. Assume that the cells can be both numbers or strings (you should plan a conversion strategy for statistical data analysis).

```python
class DataFile(object):

  def __init__(self, filename='undef'):
    with open(filename) as f:
      header = f.readline().rstrip()
      self.colnames = header.split(";")
      self.coldata = [ [] for _ in self.colnames]
      for line in f:
        line = line.rstrip()
        vals_from_file = line.split(";")
        for idx, data_column in enumerate(self.coldata):
          try:
            data_column.append( float(vals_from_file[idx]) )
          except:
            data_column.append(vals_from_file[idx])


  def info(self):
    print(f"{' '*10}{'Min':^10} {'Max':^10} {'Avg':^10}{'Distinct':^10}")
    for i,c in enumerate(self.colnames):
      print(f"{c+': ':<8}: {self.min(colnum=i):^9}{self.max(colnum=i):^13}{self.avg(colnum=i):^10.5}{self.distinct(colnum=i):^10}")
    
  def avg(self, colnum=0, colname=''):
    """ The column name or colnum can be provided alternatively
    """
    s = 0
    c = 0
    for v in self.coldata[colnum]:
      try:
        s += v
        c += 1
      except:
        pass
    return s/c if c != 0.0 else 0.0
    
  def min(self, colnum=0, colname=''):
    if colnum is not None:
      return min(self.coldata[colnum])
    return None   
      
  def max(self, colnum=0, colname=''):
    if colnum is not None:
      return max(self.coldata[colnum])
    return None    

  def distinct(self, colnum=None, colname=None):
    "Counts distinct number of values in a given column."
    if colnum is not None:
      return len(set(self.coldata[colnum]))
    return 0 

df = DataFile("mydata.txt")
df.info()
```
